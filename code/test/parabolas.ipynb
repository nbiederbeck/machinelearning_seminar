```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def func(x, a, b, x1, x2):
    return a * (x - x1) ** 3 + b * (x - x2) ** 2
```

```python
def func_(x, a, b, x1, x2):
    return 3 * a * (x - x1) ** 2 + 2 * b * (x - x2)
```

```python
def nst(a, b, x1, x2):
    return (x1 - (2 * b / (3 * a))) / 2 + np.sqrt((x1 - (2 * b / (3 * a))) ** 4 / 4 + x1 ** 2 + (2 * b) / (3 * a) * x2)
```

```python
a = -20
b = 20
x1 = 0.1
x2 = x1
```

```python
plt.plot(x, func(x, a, b, x1, x2))
plt.scatter(0, nst(a, b, x1, x2))
```

```python
plt.plot(x, func_(x, a, b, x1, x2))
```
