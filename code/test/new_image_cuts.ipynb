```python
from mls.rainbow.cutimage import *
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image as pil
%matplotlib inline
```

```python
path = '../../../../../Pictures/bliss.jpg'
path0 = './colorcircle2.jpg'

im = np.array(pil.open(path0))

plt.imshow(im)
plt.show()

theta = -np.pi / 2.5
scale = -2
faktor = 1.1
off = -0.4

mask = cutter.cut_function(
    im[:, :, 0],
    im[:, :, 1],
    faktor * im[:, :, 2] + off,
    faktor * im[:, :, 2] + off,
    theta, scale)

cut = im.copy()
cut[~mask] = 0

plt.imshow(cut)
plt.show();
```
