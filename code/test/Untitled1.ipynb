```python
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO
import imageio
from tqdm import tqdm
```

```python
N = 256
colors = np.linspace(0, 1, N)
```

```python
mesh = np.zeros((N, N, 3))
```

```python
r = 0
g = 1
b = 2
for i in range(N):
    mesh[i, :, r] = colors.copy()
    mesh[:, i, g] = colors.copy()
```

```python
def make_mesh(mesh, i):
    mesh[:, :, b] = (i / 255)
    return mesh
```

```python
def layer(mesh):
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.imshow(mesh, origin="lower", interpolation=None)
    ax.set_xticks([])
    ax.set_yticks(ax.get_xticks())
    fig.tight_layout(pad=0)
    return fig
```

```python
def make_gif(figures, filename, fps=10, **kwargs):
    images = []
    for fig in tqdm(figures):
        output = BytesIO()
        fig.savefig(output)
        plt.close(fig)
        output.seek(0)
        images.append(imageio.imread(output))
    imageio.mimsave(filename, images, fps=fps, **kwargs)
```

```python
snapshots = (layer(make_mesh(mesh, i)) for i in range(N))

make_gif(snapshots, "/tmp/colors.gif", fps=50)
```
