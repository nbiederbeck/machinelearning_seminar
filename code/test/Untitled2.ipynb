```python
from mls.rainbow.cutimage import *
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image as pil

%matplotlib inline
```

```python
cutter = RainbowCutter(
    N=265,
    theta=-np.pi/2.5,
    scale=-3,
    xshift=1.3,
    yshift=1.3,
    xbias=0.0,
    ybias=0.2,
)
```

```python
path = '../../../../../Pictures/bliss.jpg'
im = np.array(pil.open(path))
plt.imshow(im)
plt.show();
cut = cutter.cut_image(im)
plt.imshow(cut)
plt.show();
```
