```python
from mls.rainbow.cutter import RainbowCutter
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image as pil

%matplotlib inline
```

```python
cutter = RainbowCutter(
    N=256,
    theta=-np.pi/2.5,
    scale=-2.8,
    xshift=1.2,
    yshift=1.2,
    xbias=0.0,
    ybias=0.2,
)
```

```python
path = '../../../../../Pictures/bliss.jpg'
im = np.array(pil.open(path), dtype=np.uint16)
im = im / 255
plt.imshow(im)
plt.show()
cut = cutter.cut_image(im)
plt.imshow(cut)
plt.savefig("/tmp/picture.png")
```
